2. Escribe un programa que reciba tres números como entrada y determine cuál de ellos es el mayor.

```python
num1 = int(input("Escribe el primer número: "))
num2 = int(input("Escribe el segundo número: "))
num3 = int(input("Escribe el tercer número: "))

if num1 > num2 and num1 > num3:
    print("El primer número es el mayor")
elif num2 > num1 and num2 > num3:
    print("El segundo número es el mayor")
else:
    print("El tercer número es el mayor")
```

4. Escribe un programa que cuente cuántas vocales hay en una cadena dada por el usuario.

```python
cadena = input("Escribe algo: ")

vocales = 0

for i in cadena:
    i = i.lower()
    if i=="a" or i=="e" or i=="i" or i=="o" or i=="u":
        vocales+=1

print("La cadena tiene",vocales, "vocales")
```

5. Escribe un programa que imprima la tabla de multiplicar de un número dado por el usuario.

```python
num_tabla = int(input("Escribe el número del que quieres la tabla de multiplicar: "))

print(f"\nTabla del {num_tabla}:")

for i in range(1, 10+1):
    print(f"{num_tabla} x {i} = {num_tabla*i}")
```

6. Escribe un programa que reciba un número y cuente cuántos divisores tiene, es decir, cuántos números del 1 al número ingresado son divisores exactos.

```python
num = int(input("Escribe un número: "))

divisores = 0

for i in range(1, num+1):
    if num%i == 0:
        divisores += 1

print(f"El número tiene {divisores} divisores")
```

7. Escribe un programa que imprima todos los números primos entre 1 y 100.

```python
for i in range (1,100+1):
    primo = True
    for j in range (1, i+1):
        if i%j==0 and j!=1 and j!=i:
            primo = False
    if primo:
        print(i)
```

8. Escribe un programa que sume todos los números impares entre un rango de dos números dados por el usuario.

```python
num1 = int(input("Escribe el primer número: "))
num2 = int(input("Escribe el segundo número: "))

if num1 > num2:
    mayor = num1
    menor = num2
else:
    mayor = num2
    menor = num1

suma = 0
for i in range(menor, mayor+1):
    if i%2==1:
        suma+=i

print(f"La suma de los números impares entre los dos números es: {suma}")
```

9. Un número perfecto es aquel que es igual a la suma de sus divisores propios. Escribe un programa que verifique si un número dado es perfecto.

```python
num_a_comprobar = int(input("Escribe el número que quieres comprobar si es perfecto: "))

suma = 0
for i in range(1, num_a_comprobar):
    if num_a_comprobar%i==0:
        suma+=i

if suma == num_a_comprobar:
    print(f"El número {num_a_comprobar} es perfecto")
else:
    print(f"El número {num_a_comprobar} no es perfecto")
```

10. Escribe un programa que imprima los primeros n términos de la secuencia de Fibonacci, donde n es ingresado por el usuario.

```python
num = int(input("Escribe un número: "))

numeros = []

for i in range(num):
    if i == 0:
        numeros.append(0)
    elif i == 1:
        numeros.append(1)
    else:
        numeros.append(numeros[i-2]+numeros[i-1])

print(numeros)
```

11. Escribe un programa que elija un número aleatorio entre 1 y 100. Luego, el usuario intentará adivinar el número y el programa debe responder si el número es mayor o menor que el intento del usuario. El programa termina cuando el usuario adivina correctamente.

```python
import random as r

num_a_adivinar = r.randrange(0, 101)
correcto = False

print("Se ha generado un número aleatorio del 1 al 100")
print("Intenta averiguarlo!")

while not correcto:
    num = int(input("Escribe un número: "))

    if num_a_adivinar == num:
        correcto = True
    elif num_a_adivinar > num:
        print("El número a adivinar es mayor >")
    else:
        print("El número a adivinar es menor <")


print(f"Acertaste, el número a adivinar era el {num_a_adivinar}")
```
